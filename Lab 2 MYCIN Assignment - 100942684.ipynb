```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def combine_certainty_factors(X, Y):
    if X < 0 and Y < 0:
        return X + Y + X * Y
    elif X > 0 and Y > 0:
        return X + Y - X * Y
    elif abs(X) == 1 and abs(Y) == 1:
        return 1
    else:
        return (X + Y) / (1 - min(abs(X), abs(Y)))

# Generate a grid of X and Y values
X = np.linspace(-1, 1, 100)
Y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(X, Y)

# Calculate Z values for combined certainty factors
Z = np.vectorize(combine_certainty_factors)(X, Y)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Combined Certainty Factor')
ax.set_title('Combined Certainty Factors for X and Y')

plt.show()
```
